### Imports
imports and key initilization

````python
import jax
import jax.numpy as jnp
from matplotlib import pyplot as plt

from tqdm import tqdm_notebook as tqdm
from matplotlib.pyplot import figure

%matplotlib inline
````

````python
key = jax.random.PRNGKey(758493)
key2 = jax.random.PRNGKey(5616)
key3 = jax.random.PRNGKey(616)
plt.rcParams["figure.figsize"] = (12,7)
````

### Generating Data

````python
m = [5,15,120,120,120,120,120,120]
X = [jax.random.uniform(key,shape=(i,),minval=0,maxval=1) for i in m]
````

### Test and Train split
Using a ratio of 90\% and 10% each

````python
ratio = 0.8

T = []
for i in range(0,len(m)):
    X[i]=jax.random.permutation(key3,X[i],independent=True)
    num = int(m[i]*ratio)
    T.append(X[i][num:])
    X[i] = X[i][:num]
````

````python
mu = 0.0
sigma_gen = jnp.sqrt(0.3)

def generate_sin(X):
    return jnp.sin(2*jnp.pi*X) + mu+sigma_gen*jax.random.normal(key2,shape=(len(X),))
````

````python
Y = [generate_sin(i) for i in X]
Y_t = [generate_sin(i) for i in T]
````

````python
plt.scatter(X[2],Y[2])
````

### Using RBF

#### Generating basis function

````python
key2,key3 = jax.random.split(key2)

# Using a variance of 0.1 for RBF 
sigma_RBF = 0.1**0.5
basis_size = [1,3,5,10,15,20,30,50]

# Using random points from X as the basis function for RBF 
basis = [jax.random.choice(key3,X[i],shape=(basis_size[i],),replace=False) for i in range(len(m))]
# basis = [jnp.linspace(0,1,(i)) for i in basis_size]
````

#### Transforming data using RBF

````python
def create_features(basis,X:float):

    # Adding an extra one column for the bias term when perfoming linear regression
    return [1] + [jnp.exp(-((X-i)**2/(2*sigma_RBF**2))) for i in basis]  
````

````python
RBF_features = list()

for i in range(len(m)):
    tmp = list()
    for j in range(len(X[i])):
        tmp.append(create_features(basis[i],X[i][j]))
    RBF_features.append(jnp.array(tmp))
````

````python
test_RBF = list()

for i in range(len(m)):
    tmp = list()
    for j in range(len(T[i])):
        tmp.append(create_features(basis[i],T[i][j]))
    test_RBF.append(jnp.array(tmp))
````

### Online Learning using Linear Regression with the help of MLE

#### Hyperparams

````python
epoch = 15
lr = 7e-4
````

#### Using deterministic method

````python
pred_opt = []

for j in range(len(m)):
    x_man = RBF_features[j]
    # print(x_man.shape,Y[j][:,None].shape)
    w_opt = jnp.linalg.inv(x_man.T@x_man)@x_man.T@Y[j][:,None]
    pred_opt.append(x_man@w_opt)
````

#### Training Using SGD

````python
def cal_loss(W,X,Y):
    tmp = jnp.mean((Y-X@W)**2)
    return tmp,tmp


def update_weight(W,X,Y,lr):
    grad_fn = jax.jit(jax.grad(cal_loss,argnums=0,has_aux=True))
    grad,loss = grad_fn(W,X,Y)
    return (W - lr*grad,loss)
````

````python
W = []
y_pred = []
loss = []
test_pred = []

for j in tqdm(range(len(m)),desc='dataset'):

    weight = jnp.ones(shape=(RBF_features[j].shape[1]))
    weight = weight[:,None]

    tmp_loss = []
    for i in tqdm(range(epoch),desc='epoch'):
        for k in range(RBF_features[j].shape[0]):
            xtrain = RBF_features[j][k,:]
            ytrain = Y[j][k]
            weight,loss_sgd = update_weight(weight,xtrain,ytrain,lr)
            tmp_loss.append(loss_sgd)
    W.append(weight)
    y_pred.append(RBF_features[j]@weight)
    test_pred.append(test_RBF[j]@weight)
    loss.append(tmp_loss)
````

#### Plotting


````python
def plot_MSE(i):
    a,b,c,d = zip(*sorted(zip(X[i],Y[i],y_pred[i],pred_opt[i])))
    plt.scatter(a,b,c='b')
    plt.plot(a,c,c='r')
    # plt.plot(a,d,c='g')
    plt.xlabel('X axis')
    plt.ylabel('Y = sin(X) + noise')
    plt.legend(['Predicted using SGD','Data'])
    plt.title(f'Number of points = {len(a)} and kernels = {RBF_features[i].shape[1]-1}')
````

````python
plt.subplot(2,2,1)
plot_MSE(0)

plt.subplot(2,2,2)
plot_MSE(1)

plt.subplot(2,2,3)
plot_MSE(2)

plt.subplots_adjust(left=0.1,
                    bottom=0.1,
                    right=0.9,
                    top=0.9,
                    wspace=0.2,
                    hspace=0.4)
````

## Online Bayesian Learning

In order to determine the posterior probablity as a function of given prior and data, we first assume that prior is gaussian

Therefore we can write the posterior as:
$$
p(w|Φ,y) = N(w|m_N,S_N)\\
$$
where
$$
m_N = S_N(S_0^{-1}m_0 + \betaΦ^{T}y)\\
S_N^{-1} = S_0^{-1} +  \betaΦ^{T}Φ
$$
Here, $\Phi$ denotes the incoming data and $y$ denotes the value of the function to be modeled

````python
sigma =  0.3**0.5
alpha =1e-3
beta = 2

means,std,sorted_x = [],[],[]
means_test,std_test,sorted_test = [],[],[]

for j in tqdm(range(len(m)),desc='dataset'):
# j=2

    # intiliazing weights for bayesian learning for dataset of different sizes
    degree =RBF_features[j].shape[1]
    w_mu, w_pr = jnp.zeros(degree), alpha * jnp.eye(degree)
    Phi = RBF_features[j]

    for i in range(RBF_features[j].shape[0]):
        w_pr_new = w_pr + beta * Phi[i,:,None] @ Phi[i,:,None].T
        w_cov = jnp.linalg.inv(w_pr_new)
        w_mu_new = w_cov @ ( w_pr @  w_mu  + beta * Phi[i,:,None] @ jnp.array([Y[j][i]]))
        w_mu = w_mu_new
        w_pr = w_pr_new

    y_means = Phi @ w_mu
    y_var = (1 / beta) + jnp.sum(Phi @ w_cov * Phi, axis=1)
    y_std = jnp.sqrt(y_var)
    xs,ys,ys2 = zip(*sorted(zip(X[j], y_means,y_std)))
    ys,ys2 = jnp.array(ys),jnp.array(ys2)
    sorted_x.append(xs)
    means.append(ys)
    std.append(ys2)

    y_means = test_RBF[j] @ w_mu
    y_var = (1 / beta) + jnp.sum(test_RBF[j] @ w_cov * test_RBF[j], axis=1)
    y_std = jnp.sqrt(y_var)
    xs,ys,ys2 = zip(*sorted(zip(T[j], y_means,y_std)))
    ys,ys2 = jnp.array(ys),jnp.array(ys2)
    sorted_test.append(xs)
    means_test.append(ys)
    std_test.append(ys2)
````

````python
def plot_bayesian(i):
    plt.scatter(X[i], Y[i], s=50, facecolor="none", edgecolor="steelblue", lw=2)
    plt.plot(sorted_x[i], means[i],color='red')
    plt.fill_between(sorted_x[i],means[i]+std[i], means[i]-std[i], color="yellow", alpha=0.5)
    # plt.xlim(0, 1)
    # plt.ylim(-2.5, 2)
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title(f'Number of data points availabe for training = {len(X[i])}')
    plt.legend(['Predictive Mean','Data'])
````

````python
plt.subplot(2,2,1)
plot_bayesian(0)

plt.subplot(2,2,2)
plot_bayesian(1)

plt.subplot(2,2,3)
plot_bayesian(2)

plt.subplots_adjust(left=0.1,
                    bottom=0.1,
                    right=0.9,
                    top=0.9,
                    wspace=0.2,
                    hspace=0.4)
````

As we can observe from the above plots and the plots of linear regression using SGD, bayesian learning good poor in case of  low sample size while as the sample size increases, the number of parameters increases, so does number of computations increases drastically. Therefore, due to computation reasons, we avoid using bayesian learning in case of large dataset

### Analysis of model complexity and number of training points using Bayesian linear regression


#### Variation of train and test MSE vs model complexity
Here, we assume model complexity to be the number of basis functions used and the given graph has been analyzed for 96 train samples and 24 test samples with same basis function used

````python
start = 2
train_MSE = []
test_MSE = []
degree = []

for i in range(start,len(m)):
    train_MSE.append(jnp.sum((Y[i] - means[i])**2))
    test_MSE.append(jnp.sum((Y_t[i] - means_test[i])**2))
    degree.append(basis_size[i])
````
